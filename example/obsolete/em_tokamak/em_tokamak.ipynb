```python
r=f1["/record/0/H"]["x"][:,0,0];
z=f1["/record/0/H"]["x"][:,0,1];
theta=f1["/record/0/H"]["x"][:,0,2];
vr=f1["/record/0/H"]["v"][:,0,0];
vz=f1["/record/0/H"]["v"][:,0,1];
vtheta=f1["/record/0/H"]["v"][:,0,2];
B0=f1["/back_ground/B0"]
rh0=f1["/back_ground/rho0"]

m_data=dict(
          particle_position_x= r*np.cos(theta),
          particle_position_y= r*np.sin(theta),
          particle_position_z= z,
          particle_velocity_x= vr*np.cos(theta)-vtheta*np.sin(theta),
          particle_velocity_y= vr*np.sin(theta)+vtheta*np.cos(theta),
          particle_velocity_z= vz,
          particle_charge=f1["/record/0/H"]["f"][:,0] ,
          sorted_idx=f1["/record/0/H"]["_tag"][:,0])

bbox=np.array([[-2.5,2.5],[-2.5,2.5],[-5,5]]) 
ds=yt.load_particles(m_data, bbox=bbox,geometry='cartesian',length_unit='m',mass_unit='kg')
```

```python
data2=dict( 
            B0=(f1["/back_ground/B0"],"Tesla"),
            rho0=f1["/back_ground/rho0"], 
            number_of_particles=r.m_global_dims_[0], 
            particle_position_x= r*np.cos(theta),
            particle_position_y= r*np.sin(theta), 
            particle_position_z= z, 
            article_velocity_x= vr*np.cos(theta)-vtheta*np.sin(theta), 
            particle_velocity_y= vr*np.sin(theta)+vtheta*np.cos(theta), 
            particle_velocity_z= vz, 
            particle_charge=f1["/record/0/H"]["f"][:,0] , 
            sorted_idx=f1["/record/0/H"]["_tag"][:,0]
)
bbox2=np.array([[1.3,2.5],[-1.5,1.5],[0,np.pi]]) 
ds2 = yt.load_uniform_grid(data2, [50,50,5] , length_unit=(1.0, "m"),   bbox=bbox2,geometry=('cylindrical',("r","z","theta")),
                           periodicity=(False,False,True))
```

```python
slc = yt.SlicePlot(ds2, 'z', 'rho0')

slc.show()
```

```python
mi, ma = ds2.all_data().quantities.extrema('rho0')
tf = yt.ColorTransferFunction((0.3*ma,ma), grey_opacity=True)
# Choose a vector representing the viewing direction.
L = [0.5, 0.5 , 1.5]
# Define the center of the camera to be the domain center
c = ds2.domain_center[0]
# Define the width of the image
W = 1.5*ds2.domain_width[0]
# Define the number of pixels to render
Npixels = 1024 

cam = ds2.camera((2.0,2,0), (0,-1,0), (2,3,np.pi), Npixels, tf, fields=['rho0'],
                north_vector=[1,0,1], steady_north=True, 
                sub_samples=15, log_fields=[False])

cam.transfer_function.map_to_colormap(mi,ma, scale=1.0, colormap='Plasma')
cam.show()
```

```python
im=cam.snapshot()
cam.draw_coordinate_vectors(im)
im.show()
```
